Q1. 여러분은 파이썬을 통해 설문조사 문항의 응답내역을 분석하게 되었습니다. 문항별 응답내용에는 하나의 응답만 할 수 있는 single choice 문제와 여러 응답을 선택할 수 있는 multiple choice문제가 있습니다. 2개를 구분하기 위해 single choice 문항 번호에 "_"를 표기하지 않기로 했습니다.

```python
# 응답 설문 문항
question = ['Q2', 'Q3', 'Q4', 'Q5', 'Q6_1', 'Q6_2', 'Q6_3', 'Q6_4', 'Q6_5', 'Q6_6',
       'Q6_7', 'Q6_8', 'Q6_9', 'Q6_10', 'Q6_11', 'Q6_12', 'Q7_1', 'Q7_2',
       'Q7_3', 'Q7_4', 'Q7_5', 'Q7_6', 'Q7_7', 'Q8', 'Q9', 'Q10_1', 'Q10_2',
       'Q10_3']
```

```python
# comprehension을 이용하여 번호 추출

single_choice = [i for i in question if '_' not in i]
print(single_choice)
```

```python
# find함수를 사용
single = []
multi = []
for answer in question :
    if answer.find('_')== -1: # 찾고자 하는 문자가 없을 시 -1반환
        single.append(answer)
    else :
        multi.append(answer)

print(*single)
print(*multi)
```

Q2. 한스 로슬링(Hans Rosling, 1948년 7월 27일 ~ 2017년 2월 7일)은 스웨덴의 의사이자 통계학자로 비영리 벤처 갭마인더 재단의 공동설립자이기도 합니다. 빅데이터를 가장 잘 활용하는 보건 통계학자로 알려져 있습니다. 베스트셀러 책인 "팩트풀니스" 저자이기도 합니다. 갭마인더 사이트에서는 연도별, 국가별 GDP와 기대수명 데이터를 제공하고 있는데, 대표적으로 파이썬 라이브러리 중 'seaborn'에서 제공되는 예제 데이터가 있습니다. 오늘은 이 데이터를 활용해 문제를 풀어보겠습니다.
다음의 데이터는 연도, 국가별 기대수명을 나타내고 있는 데이터 입니다. 2011년 부터의 연도별, 국가별 평균 기대수명을 구해주세요. (2011년도 포함되게 구합니다.)

* [필수조건] : groupby나 pivot_table을 활용합니다. groupby로 구할 때 unstack이라는 기능을 사용하여 아래와 같이 컬럼에 인덱스 값을 올려서 표기할 수 있다.
* 만약 행에는 '연도'가, 열에 '국가'가 들어있고 수치 데이터 결과값이 아래와 같다면 출력형태는 조금 달라도 괜찮다.
* 데이터는 다음의 방법으로 읽어온다.

```python
import pandas as pd
import numpy as np
df = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/healthexp.csv")
df
```

```python
df.columns
```

```python
# 2011년 이후의 자료만 추출하여 저장
df = df[df['Year']>=2011]
```

```python
# pivot_table을 이용하여 정렬
df = pd.pivot_table(df, values='Life_Expectancy', index='Year', columns='Country',aggfunc='mean')
df
```

```python
# groupby와 unstack() 사용하여 정렬
df = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/healthexp.csv")

df = df[df['Year']>=2011]

df = df.groupby(['Year', 'Country'])['Life_Expectancy'].mean().unstack(level=1)
df
```
